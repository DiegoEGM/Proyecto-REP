Graficamos las estaciones de llegada y de salida de los viajes, coloreada de acuerdo al rango de edad del usuario (los rangos se toman cada 10 anios).

```python
def anio(s):
  #Regresa el anio de una fecha, para los dos formatos que usan los datos
  p = s.find('-') 
  if p == 4:
    return int(s[0 : 4])
  p = s.find(' ')
  return int(s[p - 4 : p])
```

```python
sample=Datos.sample(n=1000)
n = len(sample)
DatosProblema1 = np.zeros((n, 3), dtype = 'd')

DatosProblema1[:, 1] = sample.iloc[:, 6].values #Origen_id (Inicio)
DatosProblema1[:, 2] = sample.iloc[:, 7].values #Fin_id (Inicio)
#Calculamos edad como anio del viaje menos anio de nacimiento, luego separamos en decadas
for i in range(n):
    #print(Datos.iloc[i, 4])
    #print(anio(Datos.iloc[i, 4]))
    DatosProblema1[i][0] = int((anio(sample.iat[i, 4]) - sample.iat[i, 3]) / 10)
DatosProblema1
```

```python
ruh_m = plt.imread('map.png')
fig, ax = plt.subplots(figsize = (8,7))
IdToPlace={nomenclatura['id'][i]:(nomenclatura['longitude'][i],nomenclatura['latitude'][i]) for i in range(len(nomenclatura))}
RangeToPlace=[[DatosProblema1[i][0],int(DatosProblema1[i][1])] for i in range(len(DatosProblema1))] 
RangeToPlace=RangeToPlace+[[DatosProblema1[i][0],int(DatosProblema1[i][2])] for i in range(len(DatosProblema1))]
#print([RangeToPlace[i][1] for i in range(len(RangeToPlace))])
ax.scatter([IdToPlace[RangeToPlace[i][1]][0]+np.random.normal(0,0.001) for i in range(len(RangeToPlace))], 
    [IdToPlace[RangeToPlace[i][1]][1]+np.random.normal(0,0.001) for i in range(len(RangeToPlace))],
           c=[RangeToPlace[i][0] for i in range(len(RangeToPlace))], zorder=1, alpha= 0.8, s=10)
ax.set_title('Grafica MiBici con rangos de edad')
ax.set_xlim(BBox[0],BBox[1])
ax.set_ylim(BBox[2],BBox[3])
ax.imshow(ruh_m, zorder=0, extent = BBox, aspect= 'equal')
```

Aplicamos KMeans a los puntos (edad, latitud, longitud), considerando todos los viajes y tanto el punto de salida como el de llegada.

```python
from sklearn.cluster import KMeans

Ages = [int((anio(sample.iat[i, 4]) - sample.iat[i, 3])) for i in range(len(DatosProblema1))]
    
IdToPlace={nomenclatura['id'][i]:(nomenclatura['longitude'][i],nomenclatura['latitude'][i]) for i in range(len(nomenclatura))}
AgeToPlace=[[Ages[i],int(DatosProblema1[i][1])] for i in range(len(DatosProblema1))] 
AgeToPlace=AgeToPlace+[[Ages[i],int(DatosProblema1[i][2])] for i in range(len(DatosProblema1))]
#print([RangeToPlace[i][1] for i in range(len(RangeToPlace))])

X = np.zeros((len(RangeToPlace), 3))
kmeans = KMeans(n_clusters = 10)
X[:, 0] = [AgeToPlace[i][0] for i in range(len(AgeToPlace))]
X[:, 1] = [IdToPlace[AgeToPlace[i][1]][0] for i in range(len(AgeToPlace))]
X[:, 2] = [IdToPlace[AgeToPlace[i][1]][1] for i in range(len(AgeToPlace))]
kmeans_labels = kmeans.fit_predict(X)
plt.scatter(X[:, 1], X[:, 2], c = kmeans_labels)
```

Aplicamos PCA a estos datos para poder visualizarlos en 2 dimensiones, y los coloreamos usando las labels obtenidas por KMeans.

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
X_std = StandardScaler().fit_transform(X)
pca = PCA(2)
pca.fit(X_std)
vals_pca = pca.transform(X_std)

def myplot(score,coeff,labels=None):
    xs = score[:,0]
    ys = score[:,1]
    n = coeff.shape[0]
    scalex = 1.0/(xs.max() - xs.min())
    scaley = 1.0/(ys.max() - ys.min())
    plt.scatter(xs * scalex,ys * scaley, c = kmeans_labels)
    for i in range(n):
        plt.arrow(0, 0, coeff[i,0], coeff[i,1],color = 'r',alpha = 0.5)
        if labels is None:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, "Var"+str(i+1), color = 'g', ha = 'center', va = 'center')
        else:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, labels[i], color = 'g', ha = 'center', va = 'center')
    plt.xlim(-1,1)
    plt.ylim(-1,1)
    plt.xlabel("PC{}".format(1))
    plt.ylabel("PC{}".format(2))
    plt.grid()

scores = pca.fit_transform(X_std)
myplot(scores[:,0:2],np.transpose(pca.components_[0:2, :]), ['Edad', 'Latidud', 'Longitud'])
```

Ahora aplicamos agrupamiento aglomerativo. 

```python
from scipy.cluster.hierarchy import dendrogram, linkage, cut_tree 
linkage_data = linkage(X, method='ward', metric='euclidean')
dendrogram(linkage_data)
plt.show()
```

```python
h_cluster_id = cut_tree(linkage_data, n_clusters=4)
```

```python
plt.scatter(vals_pca[:, 0], vals_pca[:, 1], c = h_cluster_id)
```

Los resultados utilizando ambos metodos son similares, pero no es claro que los clusters obtenidos sean en realidad clusters en los datos; observamos que los puntos estan muy concentrados en el centro del plano y al mismo tiempo cada cluster se extiende a ambos lados.

Concluimos que no existen clusters al considerar las variables edad y lugar.
