```python
nomenclatura=pd.read_csv('https://www.mibici.net/site/assets/files/1118/nomenclatura_2023_05.csv',storage_options={"User-Agent": "XY"},encoding="latin1")
#Encuentro el rango del mapa
BBox=(nomenclatura.longitude.min(),   nomenclatura.longitude.max(),      
         nomenclatura.latitude.min(), nomenclatura.latitude.max())
ruh_m = plt.imread('map.png')
fig, ax = plt.subplots(figsize = (8,7))
ax.scatter(nomenclatura.longitude, nomenclatura.latitude, zorder=1, alpha= 0.8, c='b', s=10)
ax.set_title('Grafica de estaciones de MiBici')
ax.set_xlim(BBox[0],BBox[1])
ax.set_ylim(BBox[2],BBox[3])
ax.imshow(ruh_m, zorder=0, extent = BBox, aspect= 'equal')
```

```python
def anio(s):
  #Regresa el anio de una fecha, para los dos formatos que usan los datos
  p = s.find('-') 
  if p == 4:
    return int(s[0 : 4])
  p = s.find(' ')
  return int(s[p - 4 : p])
```

```python
n = len(Datos)
DatosProblema1 = np.zeros((n, 3), dtype = 'd')

DatosProblema1[:, 1] = Datos.iloc[:, 6].values #Origen_id (Inicio)
DatosProblema1[:, 2] = Datos.iloc[:, 7].values #Fin_id (Inicio)
#Calculamos edad como anio del viaje menos anio de nacimiento, luego separamos en decadas
for i in range(n):
    #print(Datos.iloc[i, 4])
    #print(anio(Datos.iloc[i, 4]))
    DatosProblema1[i][0] = int((anio(Datos.iloc[i, 4]) - Datos.iloc[i, 3]) / 10)
DatosProblema1
```
